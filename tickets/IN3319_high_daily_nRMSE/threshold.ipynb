```python
import pandas as pd
import os

import sys
from pathlib import Path
sys.path.insert(0,str(Path(os.path.abspath('.')).parent.parent))
```

```python
ticket_path = Path.home()/"OneDrive-3E/Research/Solar/tickets/2024/IN3043_accuracy_factors_analysis"
parse_path = os.path.join(ticket_path, "IN3136_EURAC_shared_folder/data_sites/parsed")
result_path = os.path.join(ticket_path, "IN3136_EURAC_shared_folder/data_sites/parsed_removed")
```

```python
parsed_ref_hour = pd.read_csv(os.path.join(parse_path, 'parsed_reference_data_hourly.csv'), index_col=0)
parsed_mod_hour = pd.read_csv(os.path.join(parse_path, 'parsed_model_data_hourly.csv'), index_col=0)
```

```python
parsed_ref_dai = pd.read_csv(os.path.join(parse_path, 'parsed_reference_data_daily.csv'), index_col=0)
parsed_mod_dai = pd.read_csv(os.path.join(parse_path, 'parsed_model_data_daily.csv'), index_col=0)
```

```python
parsed_ref_hour.index = pd.DatetimeIndex(parsed_ref_hour.index)
parsed_mod_hour.index = pd.DatetimeIndex(parsed_mod_hour.index)

parsed_ref_dai.index = pd.DatetimeIndex(parsed_ref_dai.index)
parsed_mod_dai.index = pd.DatetimeIndex(parsed_mod_dai.index)
```

```python
parsed_ref_hour_sum = parsed_ref_hour.resample('D').sum(min_count=1)
parsed_mod_hour_sum = parsed_mod_hour.resample('D').sum(min_count=1)
```

```python
parsed_ref = pd.concat([parsed_ref_hour_sum, parsed_ref_dai], axis=1)
parsed_mod = pd.concat([parsed_mod_hour_sum, parsed_mod_dai], axis=1)
```

## column-wise

```python
parsed_ref = parsed_ref.drop('evora', axis=1)
parsed_mod = parsed_mod.drop('evora', axis=1)
```

## threshold - 1500

```python
ref_aligned, mod_aligned = parsed_ref.align(parsed_mod, fill_value=0)
    
difference = mod_aligned - ref_aligned
difference[ref_aligned==0] = float('nan')
difference[mod_aligned==0] = float('nan')
```

```python
deleted_values_per_column_15 = difference.apply(lambda x: (x > 1500).sum())
print(deleted_values_per_column_15)
```

```python
total_deleted_values = (difference > 1500).sum().sum()
print("Total number of values deleted:", total_deleted_values)
```

```python
parsed_ref[difference > 1500] = float('nan')
```
