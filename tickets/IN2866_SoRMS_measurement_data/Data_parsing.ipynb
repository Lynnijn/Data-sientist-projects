```python
import os
import json

import sys
from pathlib import Path
sys.path.insert(0,str(Path(os.path.abspath('.')).parent.parent))
import library.validation.SAURAN.reading as reading
import library.validation.SoRMS.parsing as parsing
```

```python
ticket_path_rms = "/home/lhn3e/OneDrive-3E/Research/Solar/tickets/2023/IN2866_SoRMS_measurement_data"
raw_path_rms =  os.path.join(ticket_path_rms, 'raw_data')
intermediate_path_rms =  os.path.join(ticket_path_rms, 'intermediate_data/')
```

```python
with open(os.path.join(ticket_path_rms,'metadata/data.json'), 'r') as file:
    stations_metadata_json = file.read()

stations_metadata = json.loads(stations_metadata_json)
```

```python
stations_metadata
```

```python
for station in stations_metadata:
    station_name = f"{station['station name']}"
    
    station_data = reading.read_data_for_station_csv(raw_path_rms, station_name, skiprows=0, header=0)

    if station_name in ['Bluefield State College','Elizabeth City State University','South Park Mountain Data']:
        df = parsing.parse_rms(station_data, freq='5T')
    else:
        df = parsing.parse_rms(station_data)
        
    file_name = f"{station_name}.csv"
    file_path = os.path.join(intermediate_path_rms, file_name)
    df.to_csv(file_path, index_label = 'DateTime')
    print(f"Saved {file_name}")
```
